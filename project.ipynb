```python
youtube_df.category.value_counts()
```

```python
df_pie= youtube_df.groupby('category').count()  ## to show the count of youtube channels with respect to categories
df_pie
```

```python
prcnt =( df_pie['Youtuber']/df_pie['Youtuber'].sum())*100
sns.set()
plt.figure(figsize=(15,17))
plt.pie(df_pie['Youtuber'],autopct='%1.1f%%')
labels=df_pie.index
plt.legend(labels,loc='upper right', bbox_to_anchor=(-0.1, 1.))
plt.title("Category wise channel counts")
```
